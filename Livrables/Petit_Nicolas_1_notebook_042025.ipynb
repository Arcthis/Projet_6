### Préparation des features pour la modélisation

A réaliser :
* Si ce n'est pas déjà fait, supprimer toutes les colonnes peu pertinentes pour la modélisation.
* Tracer la distribution de la cible pour vous familiariser avec l'ordre de grandeur. En cas d'outliers, mettez en place une démarche pour les supprimer.
* Débarrassez-vous des features redondantes en utilisant une matrice de corrélation.
* Réalisez différents graphiques pour comprendre le lien entre vos features et la target (boxplots, scatterplots, pairplot si votre nombre de features numériques n'est pas très élevé).
*  Séparez votre jeu de données en un Pandas DataFrame X (ensemble de feautures) et Pandas Series y (votre target).
* Si vous avez des features catégorielles, il faut les encoder pour que votre modèle fonctionne.

```python
# Suppression des outliers
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Encodage des features
def encode_features(df):
    categorical_features = df.select_dtypes(include=['object']).columns
    le = LabelEncoder()
    for col in categorical_features:
        if df[col].nunique() < 10:
            df[col] = le.fit_transform(df[col])
        else:
            df = pd.get_dummies(df, columns=[col], drop_first=True)
    return df

# Standardisation des features
#def standardize_features(df):
#    scaler = StandardScaler()
#    numerical_features = df.select_dtypes(include=[np.number]).columns
#    df[numerical_features] = scaler.fit_transform(df[numerical_features])
#    return df

# Suppression des colonnes peu pertinentes
columns_to_keep = [
    'YearBuilt', 'BuildingAge', 'BuildingType', 'NumberofFloors', 'PropertyGFATotal', 'PropertyGFAParking',
    'SiteEnergyUse(kBtu)', 'TotalGHGEmissions', 'Outlier'
]
building_consumption = building_consumption[columns_to_keep]

# Suppression de la colonne YearBuilt
building_consumption = building_consumption.drop(['YearBuilt'], axis=1)

# Suppression des outliers
building_consumption_no_outliers = remove_outliers(building_consumption, 'SiteEnergyUse(kBtu)')

# Encodage des features
building_consumption_encoded = encode_features(building_consumption.copy())
building_consumption_no_outliers_encoded = encode_features(building_consumption_no_outliers.copy())

building_consumption_encoded = building_consumption_encoded.dropna()
building_consumption_no_outliers_encoded = building_consumption_no_outliers_encoded.dropna()

# Séparation des features (X) et du target (y)
def preprocess_data(df, target):
    x = df.drop(columns=[target])
    y = df[target]
    return x, y

x_full, y_full = preprocess_data(building_consumption_encoded, 'SiteEnergyUse(kBtu)')
x_no_outliers, y_no_outliers = preprocess_data(building_consumption_no_outliers_encoded, 'SiteEnergyUse(kBtu)')

# Split train/test
x_train_full, x_test_full, y_train_full, y_test_full = train_test_split(x_full, y_full, test_size=0.2, random_state=42)
x_train_no_outliers, x_test_no_outliers, y_train_no_outliers, y_test_no_outliers = train_test_split(x_no_outliers, y_no_outliers, test_size=0.2, random_state=42)

feature_names = x_train_no_outliers.columns  # Sauvegarde les noms des colonnes

scaler = StandardScaler()

x_train_no_outliers = scaler.fit_transform(x_train_no_outliers)
x_test_no_outliers = scaler.transform(x_test_no_outliers)

with open("scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)

# Convertir les numpy arrays en DataFrame pour récupérer les noms des colonnes
x_train_no_outliers = pd.DataFrame(x_train_no_outliers, columns=feature_names)
x_test_no_outliers = pd.DataFrame(x_test_no_outliers, columns=feature_names)

# Visualisation de la matrice de corrélation après standardisation
plt.figure(figsize=(10, 8))
sns.heatmap(building_consumption_encoded.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Matrice de Corrélation")
plt.show()

# Graphiques (boxplot et scatterplot) sur les données standardisées
categorical_features = ['BuildingType', 'Outlier']
for col in categorical_features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=building_consumption_encoded[col], y=building_consumption_encoded['SiteEnergyUse(kBtu)'])
    plt.xticks(rotation=45)
    plt.title(f"Boxplot de {col} vs SiteEnergyUse(kBtu)")
    plt.show()

numerical_features = ['NumberofFloors', 'PropertyGFATotal', 'PropertyGFAParking', 'TotalGHGEmissions', 'BuildingAge']
for col in numerical_features:
    plt.figure(figsize=(8, 4))
    sns.scatterplot(x=building_consumption_encoded[col], y=building_consumption_encoded['SiteEnergyUse(kBtu)'])
    plt.title(f"Scatterplot de {col} vs SiteEnergyUse(kBtu)")
    plt.show()

if len(numerical_features) < 6:
    sns.pairplot(building_consumption_encoded, vars=numerical_features + ['SiteEnergyUse(kBtu)'])
    plt.show()
```

### Comparaison de différents modèles supervisés

A réaliser :
* Pour chaque algorithme que vous allez tester, vous devez :
    * Réaliser au préalable une séparation en jeu d'apprentissage et jeu de test via une validation croisée.
    * Si les features quantitatives que vous souhaitez utiliser ont des ordres de grandeur très différents les uns des autres, et que vous utilisez un algorithme de regression qui est sensible à cette différence, alors il faut réaliser un scaling (normalisation) de la donnée au préalable.
    * Entrainer le modèle sur le jeu de Train
    * Prédire la cible sur la donnée de test (nous appelons cette étape, l'inférence).
    * Calculer les métriques de performance R2, MAE et RMSE sur le jeu de train et de test.
    * Interpréter les résultats pour juger de la fiabilité de l'algorithme.
* Vous pouvez choisir par exemple de tester un modèle linéaire, un modèle à base d'arbres et un modèle de type SVM
* Déterminer le modèle le plus performant parmi ceux testés.

```python
# CODE COMPARAISON DES MODELES
def evaluate_model(model, x_train, y_train, x_test, y_test, cv=5):
    """ Évalue un modèle avec validation croisée """
    cv_results = cross_validate(model, x_train, y_train, cv=cv, scoring=['r2', 'neg_mean_absolute_error'], return_train_score=True)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    results = {
        "R² Train": np.mean(cv_results['train_r2']),
        "R² Test": np.mean(cv_results['test_r2']),
        "MAE Train": -np.mean(cv_results['train_neg_mean_absolute_error']),
        "MAE Test": mean_absolute_error(y_test, y_pred),
    }
    return results

models = {
    "Linear Regression": LinearRegression(),
    "SVR": SVR(),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
}

results_full = {}
results_no_outliers = {}

for model_name, model in models.items():
    results_full[model_name] = evaluate_model(model, x_train_full, y_train_full, x_test_full, y_test_full)
    results_no_outliers[model_name] = evaluate_model(model, x_train_no_outliers, y_train_no_outliers, x_test_no_outliers, y_test_no_outliers)

print("\n Résultats AVEC Outliers :")
display(pd.DataFrame(results_full).T)
print("\n Résultats SANS Outliers :")
display(pd.DataFrame(results_no_outliers).T)
```

### Optimisation et interprétation du modèle

A réaliser :
* Reprennez le meilleur algorithme que vous avez sécurisé via l'étape précédente, et réalisez une GridSearch de petite taille sur au moins 3 hyperparamètres.
* Si le meilleur modèle fait partie de la famille des modèles à arbres (RandomForest, GradientBoosting) alors utilisez la fonctionnalité feature importance pour identifier les features les plus impactantes sur la performance du modèle. Sinon, utilisez la méthode Permutation Importance de sklearn. 

```python
# CODE OPTIMISATION ET INTERPRETATION DU MODELE
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

grid_search = GridSearchCV(
    RandomForestRegressor(random_state=42),
    param_grid,
    cv=5,
    scoring='r2',
    n_jobs=-1,
    verbose=1
)

grid_search.fit(x_train_no_outliers, y_train_no_outliers)

best_params = grid_search.best_params_
display("Meilleurs paramètres :", best_params)

best_model = grid_search.best_estimator_
y_pred = best_model.predict(x_test_no_outliers)
r2_best = r2_score(y_test_no_outliers, y_pred)
mae_best = mean_absolute_error(y_test_no_outliers, y_pred)

display(f"Meilleur modèle - R² Test: {r2_best:.4f}, MAE Test: {mae_best:.4f}")

if isinstance(best_model, RandomForestRegressor):
    feature_importances = best_model.feature_importances_
    feature_names = x_train_no_outliers.columns

    feature_importance_df = pd.DataFrame({
        'Feature': feature_names,
        'Importance': feature_importances
    }).sort_values(by='Importance', ascending=False)

    plt.figure(figsize=(10, 6))
    sns.barplot(x='Importance', y='Feature', data=feature_importance_df.head(10))
    plt.title("Top 10 Features les plus importantes (Random Forest)")
    plt.show()

else:
    from sklearn.inspection import permutation_importance

    perm_importance = permutation_importance(best_model, x_test_no_outliers, y_test_no_outliers, scoring='r2', n_repeats=10, random_state=42)

    feature_importance_df = pd.DataFrame({
        'Feature': x_test_no_outliers.columns,
        'Importance': perm_importance.importances_mean
    }).sort_values(by='Importance', ascending=False)

    plt.figure(figsize=(10, 6))
    sns.barplot(x='Importance', y='Feature', data=feature_importance_df.head(10))
    plt.title("Top 10 Features les plus importantes (Permutation Importance)")
    plt.show()

models_optimized = {
    "Linear Regression": LinearRegression(),
    "SVR": SVR(),
    "Random Forest (Optimisé)": best_model
}

results_optimized = {}

for model_name, model in models_optimized.items():
    results_optimized[model_name] = evaluate_model(model, x_train_no_outliers, y_train_no_outliers, x_test_no_outliers, y_test_no_outliers)

display("\n Résultats après optimisation :")
display(pd.DataFrame(results_optimized).T)
```
